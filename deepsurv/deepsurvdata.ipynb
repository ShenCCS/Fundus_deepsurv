身高、體重、BMI中位數計算 & 補值

```python
height = dataset['身高']
weight = dataset['體重']
BMI = dataset['BMI']

height_value = height.median()    # 160
weight_value = weight.median()  # 66.4
BMI_value = BMI.median()               # 25.78

height.fillna(height_value,inplace=True)
weight.fillna(weight_value,inplace=True)
BMI.fillna(BMI_value,inplace=True)
```

病歷號替換成檔名

```python
file_list = os.listdir(Image_data)
new_dataframe = pd.DataFrame(columns=dataset.columns)

for filename in file_list:
    start = 0

    for i in range(start,len(filename)):
        if filename[i] == '-':
            found = True
            break
    
    if found:
        match = False

        for j in range(0, len(dataset['病歷號'])):
            if filename[start:i] == str(dataset['病歷號'][j]):
                new_row = dataset.iloc[j]
                new_dataframe = new_dataframe.append(new_row,ignore_index = True)
                match = True
                
        if not match:
                    print(f"Not found : {filename[start:i]}")
```

```python
import joblib
import numpy as np
import pandas as pd

from skimage.io import imread 
from skimage.transform import resize
```

```python
loaded_model = joblib.load('/home/john/network/RETFound/RETFound_MAE-main/Fundus_deepsurv/Regression/xgb_model.pkl')
```

```python
Image_path_predict = "/home/john/network/RETFound/data/ROI/"
```

```python
def extract_features_predict(fundusimage):
    fundusimage = Image_path_predict + fundusimage

    image = imread(fundusimage)
    image = resize(image, (224, 224)) 

    feature_vector = image.flatten()
    
    return feature_vector
```

```python
df = pd.DataFrame(columns=['File', 'Predicted_Result'])

for fundusimage in os.listdir(Image_path_predict):
    feature = extract_features_predict(fundusimage)
    prediction = loaded_model.predict([feature])[0]
    
    df = df.append({'File': fundusimage, 'Predicted_Result': prediction}, ignore_index=True)
```

新增time欄位

```python
# AC 欄位 = 中風日期 
# AD 欄位 = 心肌哽塞日期
    
# NaN欄位直接移除
# 值為1的會再去確認

# IF 沒有中風 , time = 20230616 - 第一張眼底鏡日期
# IF 中風          , time = 發病的日期-第一張眼底鏡日期
```

```python
merged_df['病歷號'][1][-12:-4]
```

```python
from datetime import datetime
```

```python
merged_df['Time'] = ""

for i in range(len(merged_df['病歷號'])):
    merged_df['Event'][i] =  str(merged_df['Event'][i])

    if merged_df['Event'][i] != '0':

        if '(ICH)' in merged_df['Event'][i]:
            merged_df['Event'][i] = merged_df['Event'][i].replace('(ICH)', '')

        date = str(merged_df['Event'][i])
        date = datetime.strptime(date, '%Y%m%d')

        Imagedate = merged_df['病歷號'][i][-12:-4]
        Imagedate = datetime.strptime(Imagedate, '%Y%m%d')

        time = date - Imagedate
        time = abs(time.days)

        merged_df["Time"][i] = time
        
    else:

        NoStrokeDate = '20230616'
        NoStrokeDate = datetime.strptime(NoStrokeDate, '%Y%m%d')

        Imagedate = merged_df['病歷號'][i][-12:-4]
        Imagedate = datetime.strptime(Imagedate, '%Y%m%d')

        time = NoStrokeDate - Imagedate
        time = abs(time.days)
        merged_df["Time"][i] = time
```
