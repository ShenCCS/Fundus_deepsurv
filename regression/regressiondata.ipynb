# Training data csv

```python
import os
import shutil

Training_file =  '/home/john/network/RETFound/data/dataset_split/Training'
Regression_TrainingFile = '/home/john/network/RETFound/data/RegressionTrainingData/'

for root,sub,files in os.walk(Training_file):
    for file in files:
        source_file_path = os.path.join(root,file)
        destination_file_path = os.path.join(Regression_TrainingFile,file)
        shutil.copy2(source_file_path,destination_file_path)

print("Files copied to RegressionTrainingData.")
```

```python
Regression_dataset = "/home/john/network/RETFound/data/RegressionTrainingData/"
dataset = pd.read_csv("/home/john/network/RETFound/RETFound_MAE-main/Fundus_deepsurv/Regression/0131.csv")
```

```python
dataset
```

```python
file_list = os.listdir(Regression_dataset)
new_dataframe = pd.DataFrame(columns=dataset.columns)

for filename in file_list:
    start = 0

    for i in range(start,len(filename)):
        if filename[i] == '-':
            found = True
            break
    
    if found:
        match = False

        for j in range(0, len(dataset['病歷號'])):
            if filename[start:i] == str(dataset['病歷號'][j]):
                new_row = dataset.iloc[j]
                new_dataframe = new_dataframe.append(new_row,ignore_index = True)
                match = True
                
        if not match:
                    print(f"Not found : {filename[start:i]}")

new_dataframe.to_csv("/home/john/network/RETFound/data/Regression_TrainingData.csv",index = False)      
```
